```python
import matplotlib.pyplot as plt
%matplotlib inline

def decorateGraph():
    plt.xlabel("x-values")
    plt.ylabel("y-values")
    plt.title("Interesting graph")
    plt.legend()
```

```python
import csv 
```

```python
x= []
y= []

with open('example.csv', 'r') as csvfile:
    plots= csv.reader(csvfile, delimiter= ',')
    for row in plots:
        x.append(int(row[0]))
        y.append(int(row[1]))
        
plt.plot(x,y, label= "Loaded from file")

decorateGraph()
```
