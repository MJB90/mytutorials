```python
apl= pd.read_csv('apple_stockPrices_y2k_to_latest.csv', index_col='Date', parse_dates=True)
a5= apl.copy()
a5['H-L']= a5['High']-a5['Low']
a5['close_100ma']= pd.rolling_mean(a5['Close'], 100, min_periods=1)
a5['open_diff']= a5['Open'].diff()
a5['diff_cat']= a5['open_diff'].map(lambda x: 'increase' if x>0 else 'decrease' if x<0 else 'no_change')
a5['close_25std']= pd.rolling_std(a5['Close'], 25, min_periods=1)
a5.head()
```
